```python
data=pd.Series([1, '2', '3', 'jin', True, 10.1])
```

```python
#숫자로 변경할 수 없는 데이터라면 숫자로 변경하지 않고 원본 데이터를 그대로 반환
pd.to_numeric(data, errors='ignore')
```

```python
#숫자로 변경할 수 없는 데이터라면 기존 데이터를 지우고 NaN으로 설정하여 반환
pd.to_numeric(data, errors='coerce')
```

```python
#숫자로 변경할 수 없는 데이터라면 에러를 일으키며 코드를 중단
pd.to_numeric(data, errors='raise')
```
